```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
first_set = pd.read_csv('first_set.csv')
```

```python
first_set
```

```python
%matplotlib inline
```

```python
first_set_ra = np.zeros(len(first_set))
first_set_dec = np.zeros(len(first_set))
```

```python
first_set_ra = first_set['ra']
first_set_dec = first_set['dec']
```

```python
first_set_ra = (first_set_ra-180)*np.pi/180
first_set_dec = first_set_dec*np.pi/180
```

```python
plt.subplot(111,projection='aitoff')
plt.scatter(first_set_ra,first_set_dec)
plt.grid(True)
```

```python
first_set_u, first_set_g, first_set_r, first_set_i, first_set_z = np.zeros(len(first_set)), np.zeros(len(first_set)), np.zeros(len(first_set)), np.zeros(len(first_set)), np.zeros(len(first_set))
```

```python
first_set_u_g, first_set_g_r, first_set_r_i, first_set_i_z = np.zeros(len(first_set)), np.zeros(len(first_set)), np.zeros(len(first_set)), np.zeros(len(first_set))
```

```python
first_set_u, first_set_g, first_set_r, first_set_i, first_set_z = first_set['modelMag_u'], first_set['modelMag_g'], first_set['modelMag_r'], first_set['modelMag_i'], first_set['modelMag_z']
```

```python
first_set_u_g, first_set_g_r, first_set_r_i, first_set_i_z = first_set_u - first_set_g, first_set_g - first_set_r, first_set_r - first_set_i, first_set_i - first_set_z
```

```python
redshift = np.zeros(len(first_set))
redshift = first_set['z']
```

```python
plt.subplot(221)
plt.scatter(redshift,first_set_u_g)
plt.ylabel('u-g color')
plt.subplot(222)
plt.scatter(redshift,first_set_g_r)
plt.ylabel('g-r color')
plt.subplot(223)
plt.scatter(redshift,first_set_r_i)
plt.ylabel('r-i color')
plt.xlabel('redshift')
plt.subplot(224)
plt.scatter(redshift,first_set_i_z)
plt.ylabel('i-z color')
plt.xlabel('redshift')
```
