```python
fl = [line.split() for line in open('LC_3447107.dat')]
# as above, we are loading data from the file into a
# multi-dimensional list, where every line is a list
# where the list has values from the different columns
```

```python
fl[0:5]
# as you can see, the data is ordered
# 0 + 5*i rows contain r magnitude data
# similarly, j + 5*i rows contain time series data of the same magnitude
```

```python
rlist = []
# create an empty list to store r magnitude
for i in range(len(fl)/5):
    #print fl[0+5*i][2]
    rlist.append(fl[0+5*i][2])
    # asequentially, dd r magnitude values to the list mentioned above
```

```python
# similar to what we did above
ilist = []
for i in range(len(fl)/5):
    #print fl[1+5*i][2]
    ilist.append(fl[1+5*i][2])
```

```python
ulist = []
for i in range(len(fl)/5):
    ulist.append(fl[2+5*i][2])
glist = []
for i in range(len(fl)/5):
    glist.append(fl[3+5*i][2])
zlist = []
for i in range(len(fl)/5):
    zlist.append(fl[4+5*i][2])
```

```python
#plt.hold()
plt.plot(rlist,'.',ilist,'o',ulist, '.', glist, '.', zlist, '.')
plt.ylim([18,22])
```
